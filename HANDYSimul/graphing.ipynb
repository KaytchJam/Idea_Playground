```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read in the data
df = pd.read_csv('output/simulation.csv')
df.head()
```

```python
# Plot the data
TIME = df['timestep']
ELITES = df['elites']
COMMONERS = df['commoners']
NATURE = df['nature']
WEALTH = df['wealth']

plt.plot(TIME, ELITES, label="elites")
plt.plot(TIME, COMMONERS, label="commoners")
plt.plot(TIME, NATURE, label="nature")
plt.plot(TIME, WEALTH, label="wealth")

plt.xlabel('timestep')
plt.ylabel('stock sizes')

plt.legend()
plt.show()
```

```python
figure, axis = plt.subplots(2, 2)

# Subplots
axis[0, 0].plot(TIME, ELITES, label="elites", color='red')
axis[0, 0].set_title("elites")
axis[0, 1].plot(TIME, COMMONERS, label="commoners", color='blue')
axis[0, 1].set_title("commoners")
axis[1, 0].plot(TIME, NATURE, label="nature", color='green')
axis[1, 0].set_title("nature")
axis[1, 1].plot(TIME, WEALTH, label="wealth", color='orange')
axis[1, 1].set_title("wealth")

plt.show()
```
